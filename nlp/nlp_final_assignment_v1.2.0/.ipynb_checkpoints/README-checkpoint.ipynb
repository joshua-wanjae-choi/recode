# nlp_final_assignment_v1.2

## 0. 설명
Word Embedding 방법들(fastext, glove, word2vec)의 성능 평가  
5개사 신문들(조선, 동아, 한겨례, 중앙, 경향)로 학습진행(신문 파일은 미첨부)  

## 1. 환경
python 3.6  
관련 package 설치법은 "package" 설치 폴더를 참고  

## 2. 파일구조
```
 data(dir)  
 |  
 |___news(dir)  
 |   |  
 |   |___chosun_full.txt, donga_full.txt, hani_full.txt, joongang_full.txt,
 |        kh_full.txt  
 |  
 |  
 model(dir)  
 |   |  
 |   |___fasttext(dir)  
 |   |___glove   (dir)  
 |   |___word2vec(dir)  
 |  
 |  
 output(dir)  
 |  
 |  
 package 설치(dir)  
 |   |  
 |   |___konlpy_mecab_설치(line_encoding_ubuntu).txt, model설치.txt  
 |  
 |___train_model.sh, load_model.sh, save_whole.sentences.py, train_model.py  
```


## 3. 실행
최초 시작 시, 터미널에서 bash train_model.sh 입력  
동일 모델 재시작 시, 터미널에서 bash load_model.sh 입력  


__*hyper-parameter를 바꿔가며 model을 훈련하고 싶은 경우__  
bash train_model.sh에 적힌  
 `python train_model.py --size 100 200 300 --iter 5 10 --sg 0 1`  
에서 원하는 hyper-parameter 뒤에 숫자를 바꿔주면 됨(여러개 입력 시, 한 칸 띄어쓰기 후 입력)  
예를 들면, size를 바꾸고 싶다면 --size 100 300 500 로 바꿔줌  



## 4. 결과
output은 output 폴더 내에 csv 파일로 저장됨  
csv를 열 수 있는 응용프로그램으로 열어서 확인
